```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
x = 6 * np.random.rand(100,1) - 3
y = 0.5 * x**2 + 1.5*x + 2 + np.random.randn(100,1)    #Quadratic Equatio used
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.xlabel('X Points')
plt.ylabel('Y Points')
plt.show()
```

Building the model

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
reg_model1=LinearRegression()
```

```python
reg_model1.fit(x_train,y_train)
```

```python
score1=r2_score(y_test,reg_model1.predict(x_test))
print(score1)
```

the accuracy will be different for each time run since it is randomly generated.

```python
plt.plot(x_train,reg_model1.predict(x_train), color='r')
plt.scatter(x_train,y_train)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

Linear regression doesn't work here properly. Large difference btwn the predicted and observed values.

### Polynomial Regression

```python
p=PolynomialFeatures(degree=2, include_bias=True)
x_train_p = p.fit_transform(x_train)
x_test_p = p.transform(x_test)
```

```python
x_train_p
```

```python
x_test_p
```

```python
reg_model2=LinearRegression()
reg_model2.fit(x_train_p,y_train)
y_predict2=reg_model2.predict(x_test_p)
score2 = r2_score(y_test,y_predict2)
```

```python
print(score2)
```

```python
print(reg_model2.coef_)
print(reg_model2.intercept_)
```

```python
#degree - 3
p=PolynomialFeatures(degree=3, include_bias=True)
x_train_p3 = p.fit_transform(x_train)
x_test_p3 = p.transform(x_test)

reg_model3=LinearRegression()
reg_model3.fit(x_train_p3,y_train)
y_predict3=reg_model3.predict(x_test_p3)
score3 = r2_score(y_test,y_predict3)
print(score3)
```

```python
#degree - 4
p=PolynomialFeatures(degree=4, include_bias=True)
x_train_p4 = p.fit_transform(x_train)
x_test_p4 = p.transform(x_test)

reg_model4=LinearRegression()
reg_model4.fit(x_train_p4,y_train)
y_predict4=reg_model4.predict(x_test_p4)
score4 = r2_score(y_test,y_predict4)
print(score4)
```

```python
plt.scatter(x_train,reg_model2.predict(x_train_p))
plt.scatter(x_train,y_train)
```
