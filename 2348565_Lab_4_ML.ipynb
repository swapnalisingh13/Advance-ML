```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/Program_4_Dataset.csv")
df.head()
```

There are 5 columns in the dataset.

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
sns.pairplot(df)
plt.show()
```

Building the models

```python
x = df[['Temperature']]
y = df[['Pressure']]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.show()
```

### Train test Split

### **80-20 ratio**

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

Degree 2

```python
p=PolynomialFeatures(degree=2, include_bias=True)
x_train_p1 = p.fit_transform(x_train)
x_test_p1 = p.transform(x_test)
```

```python
reg_model1=LinearRegression()
reg_model1.fit(x_train_p1,y_train)
y_predict1=reg_model1.predict(x_test_p1)
score1 = r2_score(y_test,y_predict1)
```

```python
print('R2 score: ',score1)
mse1 = mean_squared_error(y_test, y_predict1)
print('MSE: ',mse1)
mae1 = mean_absolute_error(y_test, y_predict1)
print('MAE: ',mae1)
```

```python
plt.scatter(x_train,reg_model1.predict(x_train_p1))
plt.scatter(x_train,y_train)
```
